# Toplulaştırma ve Gruplama # 

```python
# Kadınların ve erkeklerin yaş ortalamsına erişmek istiyoruz.
df.groupby("sex")["age"].mean()
```

```python
# sadece ortalaması değil toplamı da alınmak isteniyor.
df.groupby("sex").agg({"age": ["mean", "sum"]})
```

```python
# cinsiyete göre kırdıktan sonra, gemiye biniş limanını ifade
# eden bu değişkenle ilgili de frekans bilgisi istiyoruz.
```

```python
df.groupby("sex").agg({"age": ["mean", "sum"],
                      "survived": "mean"})
```

```python
df.groupby(["sex", "embark_town", "class"]).agg({"age": ["mean", "sum"],
                                      "survived": ["mean", "count"]})
```

```python
df.pivot_table("survived", "sex", "embarked")
```

```python
df.pivot_table("survived", "sex", "embarked", aggfunc="std")
```

```python
df.pivot_table("survived", "sex", ["embarked", "class"])
```

****MERGE İLE BİRLEŞTİRME İŞLMELERİ****

```python
df1 = pd.DataFrame({'employees': ['john', "dennis", 'mark', 'maria'],
                      'group': ['accounting', 'engineering','engineering', 'hr']})

df2 = pd.DataFrame({"employees": ['mark', 'john', "dennis", 'maria'],
                       'start_date': [2010, 2009, 2014, 2019]})
```

```python
df1.head()
df2.head()
```

```python
pd.merge(df1, df2, on="employees")
```

```python
# Her çalışanın müdür bilgisine erişmek istiyoruz #
df3 = pd.merge(df1, df2)
df4 = pd.DataFrame({'group': ['accounting', 'engineering', 'hr'],
                     'manager': ['Caner', 'Mustafa', 'Berkcan']})
```

```python
df4.head()
```

```python
pd.merge(df3, df4)
```
